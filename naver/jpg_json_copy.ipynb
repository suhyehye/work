#### jpg,json 이름 똑같은 것만 복사

```python
jpg_files = glob.glob(r'C:\Users\alchera\Desktop\Naver_landmark\38-points_v2\*.jpg')
print(jpg_files)
```

```python
json_files = glob.glob(r'C:\Users\alchera\Desktop\Naver_landmark\38-points_v2\*.json')
```

```python
for jpg_file in jpg_files:
    jpg_name = jpg_file.split('\\')[-1][:-4]

    for json_file in json_files:
        json_name = json_file.split('\\')[-1][:-13]

        if json_name == jpg_name:
            print(json_name)
            shutil.copy(jpg_file,r'C:\Users\alchera\Desktop\Naver_landmark\38-points-v2_1')
            shutil.copy(json_file,r'C:\Users\alchera\Desktop\Naver_landmark\38-points-v2_1')
```
